```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch data
ticker = "RELIANCE.NS"
data = yf.download(ticker, period="6mo")

# Calculate indicators
data['EMA_50'] = data['Close'].ewm(span=50, adjust=False).mean()
data['EMA_200'] = data['Close'].ewm(span=200, adjust=False).mean()
window = 20
data['SMA_20'] = data['Close'].rolling(window=window).mean()
data['STD_20'] = data['Close'].rolling(window=window).std()
data['Upper Band'] = data['SMA_20'] + (data['STD_20'] * 2)
data['Lower Band'] = data['SMA_20'] - (data['STD_20'] * 2)
ema_12 = data['Close'].ewm(span=12, adjust=False).mean()
ema_26 = data['Close'].ewm(span=26, adjust=False).mean()
data['MACD'] = ema_12 - ema_26
data['Signal Line'] = data['MACD'].ewm(span=9, adjust=False).mean()
data['MACD Histogram'] = data['MACD'] - data['Signal Line']
data['RSI'] = calculate_rsi(data, window=14)
donchian_period = 20
data['Upper Channel'] = data['High'].rolling(window=donchian_period).max()
data['Lower Channel'] = data['Low'].rolling(window=donchian_period).min()
data['Middle Channel'] = (data['Upper Channel'] + data['Lower Channel']) / 2
lwti_period = 14
data['LWTI'] = np.nan
data['LWTI'].iloc[lwti_period-1:] = calculate_lwti(data['Close'], lwti_period)
vma_period = 20
data['VMA_SMA'] = data['Volume'].rolling(window=vma_period).mean()
data['VMA_EMA'] = data['Volume'].ewm(span=vma_period, adjust=False).mean()
data = data.dropna()

# Generate Buy/Sell signals
data['Buy'] = ((data['Close'] > data['Lower Band']) & (data['Close'].shift(1) <= data['Lower Band'])) | \
              ((data['MACD'] > data['Signal Line']) & (data['MACD'].shift(1) <= data['Signal Line'])) | \
              ((data['RSI'] > 30) & (data['RSI'].shift(1) <= 30)) | \
              ((data['Close'] > data['Upper Channel']) & (data['Close'].shift(1) <= data['Upper Channel'])) | \
              ((data['Close'] > data['LWTI']) & (data['Close'].shift(1) <= data['LWTI']))

data['Sell'] = ((data['Close'] < data['Upper Band']) & (data['Close'].shift(1) >= data['Upper Band'])) | \
               ((data['MACD'] < data['Signal Line']) & (data['MACD'].shift(1) >= data['Signal Line'])) | \
               ((data['RSI'] < 70) & (data['RSI'].shift(1) >= 70)) | \
               ((data['Close'] < data['Lower Channel']) & (data['Close'].shift(1) >= data['Lower Channel'])) | \
               ((data['Close'] < data['LWTI']) & (data['Close'].shift(1) >= data['LWTI']))

# Plot EMA and Bollinger Bands with Buy/Sell Signals
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.plot(data.index, data['Upper Band'], label='Upper Band', color='cyan')
plt.plot(data.index, data['Lower Band'], label='Lower Band', color='magenta')
plt.plot(data.index, data['EMA_50'], label='EMA 50', color='red')
plt.plot(data.index, data['EMA_200'], label='EMA 200', color='green')
plt.scatter(data.index[data['Buy']], data['Close'][data['Buy']], marker='^', color='green', alpha=1, label='Buy Signal')
plt.scatter(data.index[data['Sell']], data['Close'][data['Sell']], marker='v', color='red', alpha=1, label='Sell Signal')
plt.axhline(y=0, color='black', linestyle='--')
plt.title(f'{ticker} EMA and Bollinger Bands with Buy/Sell Signals')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(loc='upper left')
plt.show()

# Plot MACD and RSI with Buy/Sell Signals
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)
ax1.plot(data.index, data['MACD'], label='MACD', color='blue')
ax1.plot(data.index, data['Signal Line'], label='Signal Line', color='red')
ax1.bar(data.index, data['MACD Histogram'], label='MACD Histogram', color='green')
ax1.axhline(y=0, color='black', linestyle='--')
ax1.scatter(data.index[data['Buy']], data['MACD'][data['Buy']], marker='^', color='green', alpha=1, label='Buy Signal')
ax1.scatter(data.index[data['Sell']], data['MACD'][data['Sell']], marker='v', color='red', alpha=1, label='Sell Signal')
ax1.set_ylabel('MACD')
ax1.legend(loc='upper left')

ax2.plot(data.index, data['RSI'], label='RSI', color='purple')
ax2.axhline(y=70, color='red', linestyle='--')
ax2.axhline(y=30, color='green', linestyle='--')
ax2.scatter(data.index[data['Buy']], data['RSI'][data['Buy']], marker='^', color='green', alpha=1, label='Buy Signal')
ax2.scatter(data.index[data['Sell']], data['RSI'][data['Sell']], marker='v', color='red', alpha=1, label='Sell Signal')
ax2.set_ylabel('RSI')
ax2.legend(loc='upper left')

plt.xlabel('Date')
plt.suptitle(f'{ticker} MACD and RSI with Buy/Sell Signals')
plt.show()

# Plot Donchian Channels with Buy/Sell Signals
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.plot(data.index, data['Upper Channel'], label='Upper Channel', color='orange')
plt.plot(data.index, data['Lower Channel'], label='Lower Channel', color='brown')
plt.plot(data.index, data['Middle Channel'], label='Middle Channel', color='grey', linestyle='--')
plt.scatter(data.index[data['Buy']], data['Close'][data['Buy']], marker='^', color='green', alpha=1, label='Buy Signal')
plt.scatter(data.index[data['Sell']], data['Close'][data['Sell']], marker='v', color='red', alpha=1, label='Sell Signal')
plt.axhline(y=0, color='black', linestyle='--')
plt.title(f'{ticker} Donchian Channels with Buy/Sell Signals')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(loc='upper left')
plt.show()

# Plot LWTI with Buy/Sell Signals
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.plot(data.index, data['LWTI'], label='LWTI', color='purple')
plt.scatter(data.index[data['Buy']], data['Close'][data['Buy']], marker='^', color='green', alpha=1, label='Buy Signal')
plt.scatter(data.index[data['Sell']], data['Close'][data['Sell']], marker='v', color='red', alpha=1, label='Sell Signal')
plt.axhline(y=0, color='black', linestyle='--')
plt.title(f'{ticker} Linear Weighted Time Indicator (LWTI) with Buy/Sell Signals')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(loc='upper left')
plt.show()
```
